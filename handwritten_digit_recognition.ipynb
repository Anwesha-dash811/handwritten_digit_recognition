## Import libraries and data preparation
Importing the sklearn train_test_split library for data splitting
Importing the sklearn accuracy_score for finding and displaying accuracy score

```python
#import libraries
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#data preparation
num = len(digits.images)
data = digits.images.reshape((num, -1))
```

## Classification report, Confusion matrix, Accuracy using various algorithms
#### Support Vector Model

CONFUSION MATRIX:
This table is considered for the visualization of  the classification of a classification model. It consists of positive, negative observation of actual class and positive, negative observation of predicted class. The four observations are:
•	TP1(True Positive)
•	FN1(False Negative)
•	TN1(True Negative)
•	FP1(False Positive)
CLASSIFICATION REPORT:
This report consists of:
•	Sensitivity
•	Specificiy
•	Precision
•	Negative Predictive Value
•	Accuracy
•	FI-Score

    Sensitivity:
    This is used to calculate the true positive predictions by the total number of positive prediction. The best sensitivity
    rate is 1.0 and worst rate is 0.
        SENSITIVITY=TP1/(TP1+FN1)
    Specificity:
    This is used to calculate true negative predictons by the total number of negative prediction. The best specificity rate is
    1.0 and worst rate is 0. 
        SPECIFICITY=TN1/(TN1+FP1)
    Precision:
    This is used to calculate the true positive predictions by the predicted positives whether true or false.
        PRECISION=TP1/(TP1+FP1)
    Negative Predictive Value:
    This is used to calculate the true negative predictions by the predictions that remained negative whether  true or false.
        NPV=TN1/(TN1+FN1)
    F1-SCORE:
    It is the harmonic mean between precision and sensitivity.
        FI=((1+β^2)PRECISION.SENSITIVITY)/(β^2 (PRECISION+SENSITIVITY))
            (β is a constant which is commonly 1,2 or 0.5)
        F1= (TP1.TP1)/(TP1+TP1+FP1+FN1)
        FI=(2.TP1)/(2.TP1+FP1+FN1)
ACCURACY:
It is used to calculate the true observations to the total number of observations. The true observations are TP and TN.
    ACCURACY=(TP1+TN1)/(TP1+TN1+FP1+FN1)

## MODEL 1:SUPPORT VECTOR MODEL
This classifier is based on supervised algorithm. It is used for both classification and regression analysis. This classifier is a margin based classifier as it is used to differentiate the data between margin and hyper-plane and distinctly classify the dataset into classes. 
It deals with two group classification problems by giving the model sets for labeled type of training data for each category, it has the capability to work on text classification problem. The hard margin type of support vector model optimization problem can be solved by using the Lagrangian multiplier method

```python
from sklearn import svm
svm_classifier = svm.SVC(gamma=0.001)
svm_classifier.fit(X_train, y_train)
```

```python
predicted = svm_classifier.predict(X_test)
#for displaying the training images of the digits
_, axes = plt.subplots(2, 4)
images_and_labels = list(zip(digits.images, digits.target))
for ax, (image, label) in zip(axes[0, :], images_and_labels[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)
#for displaying the predicted images of the digits
images_and_predictions = list(zip(digits.images[num // 2:], predicted))
for ax, (image, prediction) in zip(axes[1, :], images_and_predictions[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Prediction: %i' % prediction)
```

```python
print("\nClassification report for classifier %s:\n%s\n" % (svm_classifier, metrics.classification_report(y_test, predicted)))
disp = metrics.plot_confusion_matrix(svm_classifier, X_test, y_test)
disp.figure_.suptitle("Confusion Matrix")
print("\nConfusion matrix:\n%s" % disp.confusion_matrix)
print("\nAccuracy of the Algorithm: ", svm_classifier.score(X_test, y_test))
plt.show()
```

## MODEL 2:NAIVE BAYES
This classifier is based on supervised algorithm. It is used for classification technique by Baye’s theorem to find out the probability of the attributes which don’t have any correlation with each other. All the attributes contributes independently to the probability. We can calculate the probability by building the frequency table and likelihood table. Further we need to find out the test phase from the likelihood table after the training is done. The equation for the Baye’s theorem is:
#### P(B/A)=(P(B/A).P(B))/(P(A))
#### P(B/A) =posterior probability
#### P(B)=class prior probability
#### P(A)=predictor prior probability
#### P(A/B)=predictor probability


```python
from sklearn.naive_bayes import GaussianNB
GNB_classifier = GaussianNB()
GNB_classifier.fit(X_train, y_train)
```

```python
predicted = GNB_classifier.predict(X_test)
#for displaying the training images of the digits
_, axes = plt.subplots(2, 4)
images_and_labels = list(zip(digits.images, digits.target))
for ax, (image, label) in zip(axes[0, :], images_and_labels[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)
#for displaying the predicted images of the digits   
images_and_predictions = list(zip(digits.images[num // 2:], predicted))
for ax, (image, prediction) in zip(axes[1, :], images_and_predictions[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Prediction: %i' % prediction)
```

```python
print("\nClassification report for classifier %s:\n%s\n" % (GNB_classifier, metrics.classification_report(y_test, predicted)))
disp = metrics.plot_confusion_matrix(GNB_classifier, X_test, y_test)
disp.figure_.suptitle("Confusion Matrix")
print("\nConfusion matrix:\n%s" % disp.confusion_matrix)
print("\nAccuracy of the Algorithm: ", GNB_classifier.score(X_test, y_test))
plt.show()
```

## MODEL 3:DECISION TREES
This classifier is based on classification algorithm while work on numerical as well as categorical data. It is required for creating tree shaped graph while analyzing the data. The analysis of decision trees are based on three nodes and those are root node, interior node and leaf node.
The idea behind the decision algorithm including the best attribute using information gain, gain ratio. It makes that attribute a decision tree and breaks into sub-datasets. Further, it starts building the tree and process repetition recursively[4].
#### Information gain,
#### Information(M)=-∑_(i=1)^n▒〖π log_2⁡π 〗
#### 〖Information〗_A (M)=∑_(j=1)^m▒〖M_J/M  XInformation(M_j)〗

#### Gain ratio,
#### 〖Split〗_A (M)=-∑_(j=1)^m▒〖M_J/M  log_2⁡〖M_J/M〗 〗
#### GainRatio(N)=(Gain(N))/(〖Split〗_A (M))


```python
from sklearn import tree
dt_classifier = tree.DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
```

```python
predicted = dt_classifier.predict(X_test)
#for displaying the training images of the digits
_, axes = plt.subplots(2, 4)
images_and_labels = list(zip(digits.images, digits.target))
for ax, (image, label) in zip(axes[0, :], images_and_labels[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)
#for displaying the predicted images of the digits   
images_and_predictions = list(zip(digits.images[num // 2:], predicted))
for ax, (image, prediction) in zip(axes[1, :], images_and_predictions[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Prediction: %i' % prediction)
```

## MODEL 4:RANDOM FOREST
This classifier is the ensemble learning classifier. It is used for both classification and regression analysis. It consists of a set of trees in which each tree is capable of providing a set of predictor values.
Overall, the decision trees are weak classifier and they are merged to form a random forest model. Random forest model don’t have cross validation while the other classifiers like decision tree and K-NN model have cross validation. In this classifier, more number of trees results in more accuracy. 
The logic behind this random forest classifier uses entropy, gain ratio and gini index.
#### Entropy,
#### Entropy(N)=-∑_(i=1)^n▒〖π log_2⁡π 〗

#### Gini Index,
#### Gini(N)=1-∑_(I=1)^M▒π^2 
#### 〖Gini〗_A (N)=N_1/N Gini(N_1 )+N_2/N Gini(N_2)


```python
from sklearn.ensemble import RandomForestClassifier
RF_classifier = RandomForestClassifier(max_depth=2, random_state=0)
RF_classifier.fit(X_train, y_train)
```

```python
predicted = RF_classifier.predict(X_test)
#for displaying the training images of digits
_, axes = plt.subplots(2, 4)
images_and_labels = list(zip(digits.images, digits.target))
for ax, (image, label) in zip(axes[0, :], images_and_labels[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)
#for displaying the predicted images of digits   
images_and_predictions = list(zip(digits.images[num // 2:], predicted))
for ax, (image, prediction) in zip(axes[1, :], images_and_predictions[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Prediction: %i' % prediction)
```

```python
print("\nClassification report for classifier %s:\n%s\n" % (RF_classifier, metrics.classification_report(y_test, predicted)))
disp = metrics.plot_confusion_matrix(RF_classifier, X_test, y_test)
disp.figure_.suptitle("Confusion Matrix")
print("\nConfusion matrix:\n%s" % disp.confusion_matrix)
print("\nAccuracy of the Algorithm: ", RF_classifier.score(X_test, y_test))
plt.show()
```
